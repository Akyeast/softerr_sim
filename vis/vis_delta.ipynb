```python
import matplotlib.pyplot as plt
import os
from utils import get_config, get_file_avg
```

```python
file_path = '../output/delta/'
files = sorted(os.listdir(file_path))

cfg = get_config('../cfg/delta_cfg.json')

deltas = cfg['deltas']
delta = [0.0] * len(cfg['deltas'])
```

```python
for file in files :
    if not file.endswith('.txt') :
        continue
    _params = [p.split('=') for p in file[:-4].split('&')]
    params = {k: v for k, v in _params}
    delta_avg = get_file_avg(file_path + file)[0]
    index = deltas.index(float(params['delta']))
    delta[index] = delta_avg
print(delta)
```

```python
plt.figure(figsize=(6,4))

plt.plot(deltas, delta, color='black')

### set y lim
# plt.ylim(2, 8)

plt.title('Core number per deltas')
plt.xlabel('Delta')
plt.ylabel('Core number avg')
plt.show()
```
